```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('brain_scan.png')
img.shape
```

```python
# keeping aspect ratio same 
new_w = int(img.shape[1] / 2.5)
new_h = int(img.shape[0] / 2.5)
image = cv2.resize(img, (new_w,new_h))
print(image.shape)
cv2_imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
ksize = [-1, 3]
gx = cv2.Sobel(gray,ddepth = cv2.CV_32F, dx=1, dy=0, ksize= -1)
gy = cv2.Sobel(gray,ddepth = cv2.CV_32F, dx=0, dy=1, ksize= -1)
gx = cv2.convertScaleAbs(gx)
gy = cv2.convertScaleAbs(gy)

#Sobel-> linear-> ksize = -1
# Scharr -> ksize->3
```

```python
combine = cv2.addWeighted(gx, 0.5, gy, 0.5, 0)
```
